```
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
import random
from fractions import gcd


def power(a, n, m):
  res = 1
  p = a % m
  while n:
    if (n & 1):
      res = (res * p) % m
    n >>= 1
    p = (p * p) % m
  return res

def isPrime1(n):
    if n % 2 == 0:
        return n == 2
    d = 3
    while d * d <= n and n % d != 0:
        d += 2
    return d * d > n

def isPrime2(n):
  for i in range(100):
    if n % 2 == 0:
      return n == 2
    a = random.randint(1, n - 1)
    if (gcd(a,n)!=1):
      return False
    if (power(a,n-1,n)!=1):
      return False
  return True

def gcdex(a, b):
  if b == 0:
#    print("a = ",a)
    return a, 1, 0
  else:
    d, x, y = gcdex(b, a % b)
#    print("d=",d,"x=",x,"y=",y)
    return d, y, (x - y * (a // b))

def closekey(a,m):
  x=0; y=0; g=0
  g = gcdex (a, m);
#  print("g=",g)
  if (g[0] != 1):
    return 1
  else:
    x = (g[1] % m + m) % m
    return x



print("Количество субъектов, участвующих в обмене - ")
N = int(input())
i=0
e = np.zeros(shape=(0,N))
d = np.zeros(shape=(0,N))
n = np.zeros(shape=(0,N))
p = np.zeros(shape=(0,N))
q = np.zeros(shape=(0,N))
x = np.zeros(shape=(0,N))

while (i<N):
  flagp=0; flagq=0
  #генереация ключей RSA
  print("Введите два простых числа p и q")
  pr=0
  qr=0
  while (isPrime1(pr)!=True and isPrime2(pr)!=True):
    pr=random.randint(100,5000)
  p = np.append(p,pr)
  print(pr)
  while (isPrime1(qr)!=True and isPrime2(qr)!=True):
    qr=random.randint(100,5000)
  q = np.append(q,qr)
  print(qr)

  flag=0
  n = np.append(n,p[i]*q[i])
  fi = (p[i]-1)*(q[i]-1)
  print(n[i],fi)
  while(flag==0):
    x = np.zeros(shape=(0,100))
    #выбор открытого ключа
    print("Введите открытый ключ")
    k= int(input())
    if (k<0 and k>n):
      print("Открытый ключ должен быть больше 0 и меньше n")
      flag=0
    if (flag==0 and gcd(k,fi)==1):
      print("Открытый ключ выбран верно")
      e = np.append(e,k)
      flag=1
    else:
      print("Открытый ключ выбран не верно")
      print("Предложить варианты открытого ключа? 1 - Да, 2 - Нет")
      v = int(input())
      if (v==1):
        for c in range(0, 100):
          if (gcd(c, fi) == 1):
            x = np.append(x,c)
        print(x)
        flag=0
      elif (v==2):
        flag=0

  flag=0
  #вычисляем закрытый ключ
  #d = inv(e,fi)
  d = np.append(d,closekey(e[i],fi))
 # print("d=",d)
  i=i+1

for i in range(N):
  print("e=",e[i],"n=",n[i],"d=",d[i])

s1 = np.zeros(shape=(0,N))
#алгоритм
sifr = np.zeros(shape=(0,N+1))
x = random.randint(1, 20)
for i in range(N):
  print("Ввести сообщение - ")
  s1 = np.append (s1,int(input()))
sifr = np.append(sifr,power(int(x+s1[0]),int(e[1]),int(n[1])))
print("x = ",x)
print("шифр ",sifr[0])
for i in range(1,N):
  s = power(int(sifr[i-1]),int(d[i]),int(n[i]))
  print("расшифр = ",s)
  if (N-i)==1:
    sifr = np.append(sifr,power(int(s+s1[i]),int(e[0]),int(n[0])))
    print("ce\\summa ",s+s1[i])
    print("шифр ",sifr[i])
  else:
    sifr = np.append(sifr,power(int(s+s1[i]),int(e[i+1]),int(n[i+1])))
    print("ce\\summa ",s+s1[i])
    print("шифр ",sifr[i])

s = power(int(sifr[-1]),int(d[0]),int(n[0])) - x
print("posl ",sifr[-1])
print("s ",s)

    







```
